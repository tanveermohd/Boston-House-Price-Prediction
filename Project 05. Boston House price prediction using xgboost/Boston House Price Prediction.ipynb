```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_data = pd.read_csv('boston.csv')
```

```python
house_data.head()
```

```python
house_data.rename(columns= {'MEDV':'PRICE'}, inplace = True)
```

```python
house_data.head()
```

```python
#Get the columns and rows

house_data.shape
```

```python
house_data.info()
```

```python
house_data.isnull().sum()
```

```python
#statistical measure

house_data.describe()
```

```python
correlation = house_data.corr()
```

```python
correlation
```

```python
# Constructing  a heatmap to understand the correlation
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True,annot_kws={'size':8}, cmap='Blues')
```

```python
#Splitting the data and Target

X= house_data.drop(columns='PRICE', axis=1)
Y= house_data['PRICE']
```

```python
X
```

```python
#splitting the data into training data and test data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model Training

```python
#loading the model
```

```python
model = XGBRegressor()
```

```python
#training the model

model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy for prediction on training data

training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error

score_1 = metrics.r2_score(Y_train, training_data_prediction)
print('R squared error :',score_1)
```

```python
# Mean absolute error

score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
print('Mean absolute error :',score_2)
```

Prediction on test data

```python
#accuracy for prediction on training data

test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)
print('R squared error :',score_1)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)
print('Mean absolute error :',score_2)
```

Visualizing the actual prices and prediction prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
